```python
import os
import shutil

# Helper function to restructure a breed directory
def restructure_breed_folder(base_dir, breed_name):
    class_dir = os.path.join(base_dir, breed_name)
    os.makedirs(class_dir, exist_ok=True)
    print(f"Restructuring {base_dir} for {breed_name}...")
    for item in os.listdir(base_dir):
        item_path = os.path.join(base_dir, item)
        if os.path.isfile(item_path) and item.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            shutil.move(item_path, class_dir)
    print(f"New contents of {base_dir}/ after restructuring: {os.listdir(base_dir)}")

# --- Restructure new datasets ---
restructure_breed_folder('/content/Schnauzer', 'Schnauzer')
restructure_breed_folder('/content/Shih-Tzu', 'Shih-Tzu')
restructure_breed_folder('/content/GoldenRetriever', 'GoldenRetriever')
restructure_breed_folder('/content/SiberianHusky', 'SiberianHusky')
restructure_breed_folder('/content/MiniaturePoodle', 'MiniaturePoodle')

# --- Create a unified dataset directory and move all restructured breeds into it ---
unified_dataset_root = '/content/dog_breeds_dataset'

# Clean up any previous unified dataset root to ensure a fresh start
if os.path.exists(unified_dataset_root):
    shutil.rmtree(unified_dataset_root)
os.makedirs(unified_dataset_root, exist_ok=True)

# List of all breed folders expected to be in /content/ after unzipping and initial restructuring
all_breed_folders_in_content = [
    'Beagle',
    'FrenchBulldog',
    'GermanShepherd',
    'Schnauzer',
    'Shih-Tzu',
    'GoldenRetriever',
    'SiberianHusky',
    'MiniaturePoodle'
]

# Move all breed folders into the unified root
for breed_name in all_breed_folders_in_content:
    source_path = os.path.join('/content', breed_name)
    destination_path = os.path.join(unified_dataset_root, breed_name)
    if os.path.exists(source_path):
        print(f"Moving {source_path} to {destination_path}")
        shutil.move(source_path, destination_path)
    else:
        print(f"Warning: {source_path} not found. Ensure all breed datasets are unzipped and restructured.")

print(f"Unified dataset root created at: {unified_dataset_root}")
print(f"Contents of {unified_dataset_root}: {os.listdir(unified_dataset_root)}")
```

```python
import os
import shutil
import random

# Define the unified dataset root
unified_dataset_root = '/content/dog_breeds_dataset'

# Define the directories for training and testing sets
train_split_dir = os.path.join(unified_dataset_root, 'train')
test_split_dir = os.path.join(unified_dataset_root, 'test')

# Clean up previous train/test directories if they exist from a prior run
if os.path.exists(train_split_dir):
    shutil.rmtree(train_split_dir)
if os.path.exists(test_split_dir):
    shutil.rmtree(test_split_dir)

os.makedirs(train_split_dir, exist_ok=True)
os.makedirs(test_split_dir, exist_ok=True)

# Define the split ratio (e.g., 80% train, 20% test)
split_ratio = 0.8

# Get all breed folders currently in unified_dataset_root (these are the class directories)
# Ensure we only pick up actual breed folders, not 'train'/'test' if they already exist
breed_folders = [d for d in os.listdir(unified_dataset_root)
                 if os.path.isdir(os.path.join(unified_dataset_root, d))
                 and d not in ['train', 'test']]

# Iterate through each breed folder to split its images
for breed_folder in breed_folders:
    # The actual images are inside unified_dataset_root/breed_folder/breed_folder
    original_class_path = os.path.join(unified_dataset_root, breed_folder, breed_folder)
    class_name = breed_folder # The class name is the folder name

    # Create breed subdirectories in train and test splits
    os.makedirs(os.path.join(train_split_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(test_split_dir, class_name), exist_ok=True)

    # List all images for the current breed
    images = [f for f in os.listdir(original_class_path) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
    random.shuffle(images)

    # Calculate split index
    split_idx = int(len(images) * split_ratio)

    # Move images to train and test directories
    for i, image_name in enumerate(images):
        src_path = os.path.join(original_class_path, image_name)
        if i < split_idx:
            shutil.move(src_path, os.path.join(train_split_dir, class_name, image_name))
        else:
            shutil.move(src_path, os.path.join(test_split_dir, class_name, image_name))

    # Remove the now empty original breed folder and its parent
    # (e.g., /content/dog_breeds_dataset/Beagle and /content/dog_breeds_dataset/Beagle/Beagle)
    shutil.rmtree(os.path.join(unified_dataset_root, breed_folder))

print(f"Dataset split into train ({len(os.listdir(train_split_dir))} classes) and test ({len(os.listdir(test_split_dir))} classes) directories.")

# Update the global variables for train and test directories
train_dir_new = train_split_dir
test_dir_new = test_split_dir

%store train_dir_new
%store test_dir_new
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms, models
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

train_dir = train_dir_new
test_dir = test_dir_new

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225])
])

try:
    train_data = datasets.ImageFolder(root=train_dir, transform=transform)
    test_data = datasets.ImageFolder(root=test_dir, transform=transform)
except Exception as e:
    print(f"Error loading datasets: {e}")
    print("Please ensure your dataset is structured with subdirectories for classes, e.g., 'root/class_name/image.jpg'.")
    print(f"Current train_dir: {train_dir}")
    print(f"Current test_dir: {test_dir}")
    raise

train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False)

model = models.resnet18(pretrained=True)

for param in model.parameters():
    param.requires_grad = False

num_classes = 8
model.fc = nn.Linear(model.fc.in_features, num_classes)
model = model.to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.fc.parameters(), lr=0.001)

num_epochs = 10
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {running_loss/len(train_loader):.4f}")

model.eval()
correct = 0
total = 0
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs, labels = inputs.to(device), labels.to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

print(f"Test Accuracy: {100 * correct / total:.2f}%")
```
