```python
import os
from collections import Counter
from PIL import Image
import matplotlib.pyplot as plt


MAIN_FOLDER = "Project3Data"   

sizes = []   # list of (width, height)

# Walk through ALL subfolders
for root, dirs, files in os.walk(MAIN_FOLDER):
    for file in files:
        if file.lower().endswith((".jpg", ".jpeg", ".png", ".bmp", ".gif", ".tiff")):
            path = os.path.join(root, file)
            try:
                with Image.open(path) as img:
                    sizes.append(img.size)
            except Exception as e:
                print(f"Could not open {path}: {e}")

print(f"Total images found: {len(sizes)}")


dist = Counter(sizes)
print("\n=== Combined Image Size Distribution Across All Subfolders ===")
for size, count in dist.most_common():
    print(f"{size}: {count}")



widths = [w for w, h in sizes]
heights = [h for w, h in sizes]

plt.figure(figsize=(10,5))
plt.hist(widths, bins=20)
plt.title("Distribution of Image Widths")
plt.xlabel("Pixels")
plt.ylabel("Count")
plt.show()

plt.figure(figsize=(10,5))
plt.hist(heights, bins=20)
plt.title("Distribution of Image Heights")
plt.xlabel("Pixels")
plt.ylabel("Count")
plt.show()
```
